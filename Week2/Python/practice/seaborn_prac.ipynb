```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load an example dataset provided by seaborn
tips = sns.load_dataset("tips")

# Create a scatter plot with a regression line using one line of code
sns.lmplot(x="total_bill", y="tip", data=tips)

# Display the plot
plt.show()
```

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size", 
)
plt.show()
```

```python
import seaborn.objects as so

penguins = sns.load_dataset("penguins")
(
    so.Plot(penguins, x='flipper_length_mm', y='body_mass_g', color='species', group='sex').add(so.Dot(), so.Agg("max")).label(title="Penguin's Body Mass Vs Flipper Length")
    .scale(color="flare")
)
```

```python
(
    so.Plot(penguins, x="species", y="body_mass_g", color="sex")
    .add(so.Bar(), so.Agg(), so.Dodge(), so.Jitter())
    .facet("species", wrap=3)
)
```

```python
(
    so.Plot(tips, x="total_bill", y="size", color="time")
    .add(so.Bar(), so.Agg(), so.Dodge(), orient="y")
    .scale(color="flare")
)
```

```python
diamonds = sns.load_dataset("diamonds")
(
    so.Plot(diamonds, x="carat", y="price", color="carat")
    .add(so.Dots())
    .scale(
        x=so.Continuous().tick(every=0.5),
        y=so.Continuous().label(like="${x:.0f}"),
        color=so.Continuous().tick(at=[1, 2, 3, 4]),
    )
)
```

## Types Of Seaborn Plots
- Relational Plots in Seaborn
    - Scatter Plot
    - Line Plot and 
    - Relational Plot
- Categorical Plots in Seaborn
    - Bar Plot 
    - Count Plot
    - Box Plot
    - Violin Plot
    - Strip Plot 
    - Swarm Plot
- Distribution Plots in Seaborn
    - Histogram 
    - Kernel Density Estimate Plot
    - Distribution Plot
    - Empirical Cumulative Distribution Function Plot
    - Rug Plot
- Matrix Plots in Seaborn
    - Heatmap
    - Cluster Map
- Pair Grid in Seaborn
    - Pair Plot

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
cust_err = lambda x: (x.min(), x.max())
sns.barplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    errorbar=cust_err
)
```

```python
anscombe = sns.load_dataset('anscombe')
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"), order=12, ci=None, scatter_kws={"s": 80})
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
attend = sns.load_dataset("attention").query("subject <= 12")
g = sns.FacetGrid(attend, col="subject", col_wrap=4, height=2, ylim=(0, 10))
g.map(sns.pointplot, "solutions", "score", order=[1, 2, 3], color=".3", errorbar=None)
```

```python
iris = sns.load_dataset("iris")
sns.set_style('darkgrid')
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.kdeplot)
```
