```python
# csv loader
import os
from langchain_community.document_loaders.csv_loader import CSVLoader

current_dir = os.getcwd() 

file_path = os.path.join(current_dir, "data_files", "example.csv")

if not os.path.exists(file_path):
    print(f"Error: File not found at {os.path.abspath(file_path)}")
    print(f"Current Directory is: {current_dir}")
else:
    loader = CSVLoader(
        file_path=file_path,
        csv_args={"delimiter": ","}
    )

    data = loader.load()
    print(f"Success! Loaded {len(data)} rows.")
    print(data[0:10])
```

```python
# html loader
from langchain_community.document_loaders import UnstructuredHTMLLoader, UnstructuredURLLoader
loader = UnstructuredURLLoader(urls=['https://www.geeksforgeeks.org/artificial-intelligence/langchain-document-loaders/'], mode='elements')
data_html = loader.load()

print(len(data_html))
print(data_html[0])
```

```python
# markdown
import os
import markdown
from langchain_community.document_loaders import UnstructuredMarkdownLoader

current_dir = os.getcwd()
file_path = os.path.join(current_dir, "data_files", "example.md")
loader = UnstructuredMarkdownLoader(file_path, mode='elements')
data_md = loader.load()

print(f"Loaded elements: {len(data_md)}")

print(data_md[1].page_content)
print(data_md[1].metadata)
```

```python
# pdf 
import os
from langchain_community.document_loaders import UnstructuredPDFLoader

current_dir = os.getcwd()
file_path = os.path.join(current_dir, "data_files", "sample.pdf")
if not os.path.exists(file_path):
    print(f"File not found at: {file_path}")
else:
    loader = UnstructuredPDFLoader(file_path, mode='elements', strategy="auto")
    data = loader.load()

    print(f"Success! Loaded {len(data)} elements.")
    print(data)
```

```python
# wikipedia
from langchain_community.document_loaders.wikipedia import WikipediaLoader
loader = WikipediaLoader(query='Generative AI', load_max_docs=5, doc_content_chars_max=5000, load_all_available_meta=True)
data = loader.load()

for doc in data:
    print(doc.metadata['title'])
```
