```python
number = 10
if number > 5:
    raise Exception(f"The number should not exceed 5. ({number=})")
print(number)
print("Executed Anyways!")
```

```python
# using assert
number = 1
# number = 10 # gives assertion error
assert (number < 5), f"The number should not exceed 5. ({number=})"
print(number)
```

```python
try:
    print(1/0)
except ZeroDivisionError as e:
    print(e)
else:
    print("This will executed if no error")  
finally:
    print("End of try except block")  
print("This code will run fs")
```

```python
# custom exception class 
import sys
class PlatformException(Exception):
    # pass
    def __init__(self, msg):
        self.msg = msg
        super().__init__(msg)
    def __str__(self):
        return f"{self.msg} Current Platform: {sys.platform}"

def linux_check():
    if 'linux' not in sys.platform:
        raise PlatformException("Linux required")
    print('Doing Something')

try:
    linux_check()
except PlatformException as p:
    print(p)
```

```python
str = "while True print('Hello world')"

try:
    exec(str)
except SyntaxError as e:
    print(f"{e} as SyntaxError")
except:
    pass
finally:
    print('Completed')
```

```python
# program to get a number from user
while True:
    try:
        x = int(input("Please enter a number: "))
    except ValueError:
        print("Oops!  That was no valid number.  Try again...")
    else:
        print(x)
        break
```
