```python
# decorator
def upperCase(func):
    def inner(user):
        return f"{func(user)} {user}".upper()
    return inner

@upperCase
def greet(user):
    return "Hello"
greet("Joe")
```

```python
# Generator
def fibo(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a+b

output = fibo(5)
print(f"{next(output)}  {next(output)}  {next(output)}  {next(output)}  {next(output)}".center(20))
for num in fibo(5):
    print(num)
```

```python
# python MRO
class A:
    def method(self): print("Method A")

class B(A):
    def method(self): print("Method B")

class C(A):
    def method(self): print("Method C")

class D(B, C):
    pass

print(D.__mro__)
d = D()
d.method()
```

```python
class A:
    def greet(self):
        return "Welcome to Class A"
    class B:
        def greet(self):
            return "Welcome to Class B"
# Accessing Inner Class from the Outside
b = A().B()
b.greet()

# Accessing Outer Class from the Inner
a1 = A()
b1 = a1.B()
b1.greet()
```
